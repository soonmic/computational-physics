```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def q(x,x_star):
    return (1/s) * np.exp(-(x-x_star)/s) * np.exp(-np.exp(-(x-x_star)/s))

def p(x):
    return (1/np.sqrt(2 * np.pi)) * np.exp(-(x**2)/2)

def alpha(x,x_star):
    return min(1, (p(x_star)/p(x)) * (q(x,x_star)/q(x_star,x)))
```

```python
N = 100000

x = 5 # 최초 표본

s = 0.5 # 척도

x_list = []

for i in range(N):
    x_star = np.x_star = np.random.gumbel(x, 0.5)
    x_alpha = alpha(x,x_star)
    a = np.random.uniform(0,1)
    if x_alpha >= a:
      x_list.append(x_star)
      x = x_star
    else:
      x_list.append(x)

plt.hist(x_list, bins = 100)
```
