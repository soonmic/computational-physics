### 함수

```python
def find(i):
    if ptr[i] < 0:
        return i
    else:
        ptr[i] = find(ptr[i])
        return ptr[i]

def union(r1, r2, big):
    if r2 != r1:
        if ptr[r1] > ptr[r2]:
            ptr[r2] += ptr[r1]
            ptr[r1] = r2
            r1 = r2
        else:
            ptr[r1] += ptr[r2]
            ptr[r2] = r1
        if -ptr[r1] > big:
            big = -ptr[r1]
    big_list.append(big/N)
    return r1, r2, big

def neighbors(i,j,L):
    neighbors = [(i, (j + 1) % L), (i, (j - 1) % L), ((i + 1) % L, j), ((i - 1) % L, j)] # 오,왼,아래, 위
    return neighbors
```

### 실행 코드

```python
L = 5
N = L*L

bonds = []

for i in range(L):
    for j in range(L):
        c = i * L + j
        r = i * L + (j + 1) % L
        d = ((i + 1) % L) * L + j
        bonds.append((c, r))
        bonds.append((c, d))

# np.random.shuffle(bonds)
print(bonds)
ptr = np.ones(N, dtype = int) * -1
```

```python
big = 0
big_list = []

for i in range(L**2):
  ptr[i] = -1

for bond in bonds:
    s1 , s2 = bond
    r1 = find(s1)
    r2 = find(s2)
    r1,r2,big = union(r1,r2,big)


xlist = [(i+1)/(2*N) for i in range(2*N)]

plt.plot(xlist,big_list)
plt.show()
```
