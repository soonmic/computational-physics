```python
# for문 없는 코드
import numpy as np

N = 100000
p = 0.4
n = 50

x = np.random.choice([1, -1], size=(N, n), p=[p, 1-p]) # N번의 시도를 한 번에 처리하는 코드

sums = np.sum(x, axis=1) # 누적 합계 구하는코드

probability = np.mean(sums > 15) # 15 이상인 것의 비율 계산하기

print(probability)
```

```python
import random as rd

N = 100000

p = 0.4 # 오른쪽으로 움직일 확률

total = 0 # 15이상 나온 횟수

for i in range(N) :
    x = 0
    for k in range(50) :
        p_i = rd.uniform(0,1)
        if p_i >= p :
            x -= 1
        else :
            x += 1
    if x > 15 :
        total += 1

print(total / N)
```

```python
import math

N = 100000

q = 0.65

total = 0

pq_sum = 0

for i in range(N):
    x = 0
    pq_list = []
    for k in range(50):
        p_i = rd.uniform(0,1)
        if p_i >= q: #
          x -= 1
          pq = (1-p)/(1-q)
          pq_list.append(pq)
        else:
          x += 1
          pq = p/q
          pq_list.append(pq)
    if x > 15:
      total += 1
      pq_total = math.prod(pq_list)
      pq_sum += pq_total

print(total/N)
print(pq_sum/N)
```
