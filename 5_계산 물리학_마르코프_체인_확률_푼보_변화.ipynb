```python
import numpy as np
import matplotlib.pyplot as plt

state = 'S'

total_result = []

result = []

t_list = [i for i in range(10)]

for _ in range(10000):  #   10,000 
    state = 'S'
    result = []
    for _ in range(10):
        if state == "S":
            state = np.random.choice(['S', 'C', 'R'], p=[0.7, 0.2, 0.1])
            result.append(state)
        elif state == "C":
            state = np.random.choice(['S', 'C', 'R'], p=[0.3, 0.65, 0.05])
            result.append(state)
        elif state == "R":
            state = np.random.choice(['S', 'C', 'R'], p=[0.15, 0.35, 0.5])
            result.append(state)

    total_result.append(result)

#  t  'S', 'C', 'R'    
counts_S = [0] * len(t_list)
counts_C = [0] * len(t_list)
counts_R = [0] * len(t_list)

# total_result   t  'S', 'C', 'R'  
for result in total_result:
    for t, state in enumerate(result):
        if state == 'S':
            counts_S[t] += 1
        elif state == 'C':
            counts_C[t] += 1
        elif state == 'R':
            counts_R[t] += 1

#    
total_counts = np.array(counts_S) + np.array(counts_C) + np.array(counts_R)
proportion_S = np.array(counts_S) / total_counts
proportion_C = np.array(counts_C) / total_counts
proportion_R = np.array(counts_R) / total_counts

#  
plt.plot(t_list, proportion_S, label='S')
plt.plot(t_list, proportion_C, label='C')
plt.plot(t_list, proportion_R, label='R')
plt.xlabel('Time step')
plt.ylabel('Proportion')
plt.title('Proportion of States over Time')
plt.legend()
plt.show()
```
