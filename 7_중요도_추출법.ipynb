### 예제 5.3.1

```python
import numpy as np
import random as rd

def g(x):
  return np.exp(-0.5 * (x**2))

def q(x):
  return np.exp(-x) / (1 - np.exp(-10))

def w(x):
  return 0.1 / q(x)
```

```python
N = 10000

x_i = np.random.uniform(0,10,N)

total = np.sum(g(x_i))/N

print(total)
```

```python
N = 10000

u = np.random.uniform(0,1,N)
x_i_2 = -np.log((1 - (1-(np.exp(-10))) * u))

total_2 = np.sum(g(x_i_2)*w(x_i_2))/N
print(total_2)
```
