```python
import numpy as np
import random as rd
import matplotlib.pyplot as plt
```

```python
result = []

t_list = [i for i in range(99)]

state = "S"

for i in range(99):
    num = rd.uniform(0,1)
    if state == "S":
      if num <= 0.7:
        state = "S"
        result.append(state)
      elif 0.7 < num <= 0.9:
        state = "C"
        result.append(state)
      else:
        state = "R"
        result.append(state)
    elif state == "C":
      if num <= 0.3:
        state = "S"
        result.append(state)
      elif 0.3 < num <= 0.95:
        state = "C"
        result.append(state)
      else:
        state = "R"
        result.append(state)
    elif state == "R":
      if num <= 0.15:
        state = "S"
        result.append(state)
      elif 0.15 < num <= 0.5:
        state = "C"
        result.append(state)
      else:
        state = "R"
        result.append(state)

plt.plot(t_list,result)
```
