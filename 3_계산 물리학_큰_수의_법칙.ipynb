```python
def sunmin(k):
    N = 10 ** (k+4)
    n = 10 ** k
    x_arr = np.random.uniform(0,1,N * n)
    temp = np.reshape(x_arr, (N,n))
    sum = np.sum(temp, axis=1)
    mean_sum = sum/n
    return mean_sum
```

```python
plt.hist(sunmin(1), 100) # n이 10일때
plt.hist(sunmin(2), 100) # n이 100일때
```
